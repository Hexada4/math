```python
from sympy import *
init_printing()
```

f(x)=$x^{2}$+3x-4, f'(x)=2x+3.

```python
x = Symbol('x')
f = x**2 + 3*x - 4
f
```

```python
f_ = 2*x + 3 
solution = solve(f_)[0]
solution
```

```python
plot(f, (x, -1.5, 3))
```

f(x)=$x^{3}$-$2x^{2}$+x-6, f'(x)=$3x^{2}$-4x+1.

```python
f = x**3 - 2*x**2 + x - 6
f
```

```python
f_ = 3*x**2 - 4*x + 1
f_
```

```python
solution = solve(f_)
solution
```

```python
plot(f, (x, -5, 1/3))
```

```python
plot(f, (x, 1, 5))
```

f(x)=$e^{2x-x^{2}}$, f'(x)=(2-2x)$e^{2x-x^{2}}$.


```python
f = E**(2*x - x**2)
f
```

```python
f_ = (2 - 2*x) * E**(2*x - x**2)
solution = solve(f_)
solution
```
