f(x,y)= sin($\sqrt{x^2+y^2}$) 
<!--  $\sqrt{\$4}$ -->

```python
from sympy import *
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
y = Symbol('y')
f = sin(sqrt(x**2+y**2))
f
```

```python
# распечатаем это чудо
plot3d(f)
```

\$$f(x,y)=e^-0.9x2-0.45(x-y)^2$$
. Задайте промежутки для икс и игрек (x,-3,3), (y,-3,3). Это делается точно также, как и для случая одной переменной, только в функцию plot3d вместо одного кортежа нужно передать два

```python
f = E**-0.9*x*2-0.45*(x-y)**2
f
```

```python
plot3d(f, (x, -3, 3), (y, -3, 3))
```
