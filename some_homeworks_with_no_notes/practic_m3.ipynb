```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

# 1 задание
Набор точек, для которого нужно решить задание: <b>(−2, 2), (−1, 1), (0, 0), (1, 1), (2, 2)</b>

```python
a4, a3, a2, a1, a0 = symbols('a4,a3,a2,a1,a0')
first = 16*a4 - 8*a3 + 4*a2 - 2*a1 + a0 -2
second = a4 - a3 + a2 - a1 + a0 - 1
third = a0
fourth = a4 + a3 + a2 + a1 + a0 - 1
fifth = 16*a4 + 8*a3 + 4*a2 +2*a1+a0 - 2
solves = list(nonlinsolve([first, second, third, fourth, fifth], [a4, a3, a2, a1, a0]))[0]
solves
```

```python
x = Symbol('x')
f = solves[0]*x**4+solves[1]*x**3+solves[2]*x**2+solves[3]*x+solves[4]
plot(f, (x, -3, 3))
```

```python
f.subs(x, -1)
```

```python
f.subs(x, 1)
```

```python
f.subs(x, -100)
```

```python
f.subs(x, 100) # -> чётная
```

```python
solve(f)
```

```python
solve_univariate_inequality(f>0, x)
```

```python
solve_univariate_inequality(f<0, x)
```

```python
plot(f)
```

# 2 задание
Набор точек, для которого нужно решить задание: <b>(−3, 33), (−2, 31), (−1, 18), (1, −18), (2, −31), (3, −33)</b>

```python
a5,a4,a3,a2,a1,a0 = symbols('a5,a4,a3,a2,a1,a0')
eq1 = -243*a5+81*a4-27*a3+9*a2-3*a1+a0-33
eq2= -32*a5+16*a4-8*a3+4*a2-2*a1+a0-31
eq3 = -a5+a4-a3+a2-a1+a0-18
eq4 = a5+a4+a3+a2+a1+a2+18
eq5 = 32*a5+16*a4+8*a3+4*a2+2*a1+a0+31
eq6 = 243*a5+81*a4+27*a3+9*a2+3*a1+a0+33
solves = list(nonlinsolve([eq1, eq2, eq3, eq4, eq5, eq6], [a5, a4, a3, a2, a1, a0]))[0]
solves
```

```python
f = solves[0]*x**5+solves[2]*x**3+solves[4]*x
plot(f, (x, -5, 5))
```

```python
f.subs(x, -1)
```

```python
f.subs(x, 1) # -> чётная
```

```python
solve(f)[3:]
```

```python
solve_univariate_inequality(f>0, x)
```

```python
solve_univariate_inequality(f<0, x)
```

```python
plot(f)
```
