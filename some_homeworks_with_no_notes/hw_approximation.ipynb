```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points_X = [2, 4, 8]
points_Y = [12.258, 17.24, 30.151]
x = Symbol('x')
f = 3*x+5

X = np.arange(1, 10, 0.01)
Y = [f.subs(x, x_) for x_ in X]

# (2; 12.258), (4; 17.24), (8; 30.151)
plt.scatter(points_X, points_Y, c='r')
plt.plot(X, Y, 'b')
plt.show()
```

```python
x = Symbol('x')
f1 = 3*x+5

def sigma_solver(function, points_x: list, points_y: list): 
    sum_ = 0 
    for i in range(len(points_x)):
        sum_ += (points_y[i] - function.subs(x, points_x[i]))**2
        print(sum_)    
    return (sum_/len(points_x))**(1/2)

points_Y = [12.258, 17.24, 30.151]
sigma_solver(f1, points_X, points_Y)
```

```python
points_X = [2, 4, 8]
points_Y = [3.688, 10.791, 20.705]
x = Symbol('x')
f2 = 0.25*x**2 + 0.75*x + 1.25

X = np.arange(1, 10, 0.01)
Y = [f2.subs(x, x_) for x_ in X]

# 2; 3.688), (4; 10.791), (8; 20.705).
plt.scatter(points_X, points_Y, c='r')
plt.plot(X, Y, 'b')
plt.show()
```

```python

points_Y = [3.688, 10.791, 20.705]
sigma_solver(f2, points_X, points_Y)
```
